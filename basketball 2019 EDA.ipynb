### dealing with missing values

```python
# looking missing values
mis=0
for f in df2019.columns:
    if(df2019[f].isnull().sum()>0):
        mis+=1
        print('{} has {}% missing values'.format(f,(df2019[f].isnull().sum()/df2019.shape[0])*100))
print('{} features contain missing'.format(mis))
```

```python
df2019[df2019['3P%'].isnull()]
```

```python
# based on the above table. if the value of 3p and 3pa are 0, then the value of 3p% is nan so we can fill it up with 0
df2019['3P%']=df2019['3P%'].fillna(0)
```

###### exploring some insight

```python
# creating the bar plot
def barplot(f):
    #df2019[f].value_counts().plot(kind='bar', figsize=(8, 6))
    sns.countplot(f,data=df2019,palette='hls')
    plt.xlabel("Count", labelpad=14)
    plt.ylabel("{} list".format(f), labelpad=14)
    plt.title("Count of {}".format(f), y=1.02);
    plt.xticks(rotation=90)
    plt.show()
```

```python
len(df2019['Tm'].value_counts())
```

```python
cat=df2019.select_dtypes(include='O')
for i in cat:
    barplot(i)
```

###### observation:
    1.SG,PF,PG,C,SF are popular position
    2.all teams have around 20-30 players, but TOT has more than 80 players

```python
# visualizing top 5 teams
df2019.Tm.value_counts().head(5).plot.bar(color=('blue','gray','red','green','orange','yellow','purple','violet','lime','aqua'))
plt.title("Top 5 team")
```

```python
sns.catplot(x="Pos", y="PTS",data=df2019,kind="bar",palette="muted")
plt.title("position with Point")
plt.xlabel('position')
plt.xticks(rotation=90)
plt.ylabel('points')
```

```python
sns.scatterplot(x="Pos", y="G",data=df2019,palette="muted")
plt.title("position with goal")
plt.xticks(rotation=90)
plt.xlabel('position')
plt.ylabel('goal')
```

```python
sns.violinplot(x="Pos", y="G",data=df2019,palette="muted")
plt.title("goal distribution with position")
plt.xticks(rotation=90)
plt.xlabel('position')
plt.ylabel('goal')
```
